```python
import pandas as pd
titanic_df = pd.read_csv('titanic.csv')
titanic_df
```

```python
# Check for missing values
print(titanic_df.isnull().sum())

# Fill missing 'Age' values with the median age
titanic_df['age'].fillna(titanic_df['age'].median(), inplace=True)

# Fill missing 'Embarked' values with the most common port
titanic_df['embarked'].fillna(titanic_df['embarked'].mode()[0], inplace=True)


# Convert 'Sex' and 'Embarked' to categorical variables
titanic_df['sex'] = titanic_df['sex'].astype('category')
titanic_df['embarked'] = titanic_df['embarked'].astype('category')

# Convert 'Survived' and 'Pclass' to categorical variables
titanic_df['survived'] = titanic_df['survived'].astype('category')
titanic_df['pclass'] = titanic_df['pclass'].astype('category')

# Display the updated dataframe info
print(titanic_df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib and seaborn plotting environment
plt.style.use('seaborn-darkgrid')
sns.set_palette('Set2')

# Descriptive statistics
print(titanic_df.describe(include='all'))

# Distribution of numerical variables
titanic_df.hist(bins=20, figsize=(14,10))
plt.show()

# Bar plot of survival by sex
sns.countplot(data=titanic_df, x='survived', hue='sex')
plt.title('Survival by Sex')
plt.show()

# Bar plot of survival by passenger class
sns.countplot(data=titanic_df, x='survived', hue='pclass')
plt.title('Survival by Passenger Class')
plt.show()

# Box plot of age by survival status
sns.boxplot(data=titanic_df, x='survived', y='age')
plt.title('Age by Survival Status')
plt.show()

# Pair plot to explore relationships between variables
sns.pairplot(titanic_df, hue='survived', diag_kind='kde')
plt.show()
```

```python
# Violin plot of age by survival status and passenger class
sns.violinplot(data=titanic_df, x='pclass', y='age', hue='survived', split=True)
plt.title('Age Distribution by Survival Status and Passenger Class')
plt.show()
```
